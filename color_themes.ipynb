```python
from bs4 import BeautifulSoup
from selenium import webdriver
from webdriver_manager.firefox import GeckoDriverManager
import pandas as pd
import time
import requests
from matplotlib import pyplot as plt
import matplotlib
```

```python
driver = webdriver.Firefox(executable_path=GeckoDriverManager().install())
```

```python
url = 'https://plugins.jetbrains.com/search?tags=Theme'
driver.get(url)
```

```python
themes = driver.find_elements_by_xpath('//h3[@class="mp-plugin-card__image__name wt-h3 wt-h3_theme_dark"]')
theme_list = []
for i in range(len(themes)):
    theme_list.append(themes[i].text)
theme_list
```

```python
downloads = driver.find_elements_by_xpath('//div[@class="test_plugin-downloads"]')
download_list = []
for j in range(len(downloads)):
    download_list.append(downloads[j].text)
download_list
```

```python
tool_datagrip = ['IntelliJ']*len(theme_list)
tool_datagrip
```

```python
dict = {'Tool': tool_datagrip, 'Theme': theme_list, 'Downloads': download_list}
```

```python
df = pd.DataFrame(dict)
df
```

```python
url = 'https://marketplace.visualstudio.com/search?target=VSCode&category=Themes&sortBy=Installs'
driver.get(url)
```

```python
vs_theme = driver.find_elements_by_xpath('//span[@class="item-title"]')
vs_theme_list = []
for i in range(len(vs_theme)):
    vs_theme_list.append(vs_theme[i].text)
vs_theme_list
```

```python
vs_downloads = driver.find_elements_by_xpath('//span[@class="install-count"]')
vs_download_list = []
for i in range(len(vs_downloads)):
    vs_download_list.append(vs_downloads[i].text)
vs_download_list
```

```python
tool_vscode = ['Visual Studio Code']*len(vs_download_list)
```

```python
vscode_dict = {'Tool': tool_vscode, 'Theme': vs_theme_list, 'Downloads': vs_download_list}
```

```python
df1 = pd.DataFrame(vscode_dict)
df1
```

```python
tens = {'k': 10e3, 'm': 10e6}
```

```python
df1['Downloads'] = df1['Downloads'].apply(lambda x : int(float(x[0:-1]) * tens[x[-1].lower()]))
df1
```
